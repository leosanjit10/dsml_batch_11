```python
#got from co pilot
class Calculator:
    """Simple calculator for basic operations"""
    def __init__(self, a, b):
        self.a = a   # first number (instance attribute)
        self.b = b   # second number (instance attribute)

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def multiply(self):
        return self.a * self.b

    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Error! Division by zero."

# Create an object of Calculator
c1 = Calculator(10, 5)

print("Addition:", c1.add())
print("Subtraction:", c1.subtract())
print("Multiplication:", c1.multiply())
print("Division:", c1.divide())
```

```python
class dog:
  count = 0 #class attribute
  def __init__(self,name,age=3):
    self.name = name
    self.age = age
    dog.count += 1
  def bark(self):
    print(f"{self.name} is barking")
  @classmethod
  def get_count(cls):
    return cls.count
  @staticmethod
  def get_vet_ingo():
    return "dr. ram k.c"
```

```python
dog1 = dog("puppy")
print(dog1.name)
print(dog1.age)
```
