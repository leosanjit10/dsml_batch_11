```python
def largest_number(a,b,c):
  if a>=b and a>=c:
    return a
  elif b>=a and b>=c:
    return b
  else:
    return c
print(largest_num(200,500,30))
```

```python
def vowels(text):
  vowels = "aeiouAEIOU"
  count = 0
  for char in text:
    if char in vowels:
      count+=1
  return count
print(vowels("aeroplane"))
```
