```python
#numpy methods: method to create numpy
 #array()
 #zeros()
 #ones()
 #full()
 #arrange()
 #linspace()
 #diag()
 #eye()
 #identity()
 #random.rand()
```

```python
# 1. array() → create from a Python list
arr1 = np.array([1, 2, 3, 4, 5])
print("array():", arr1)

# 2. zeros() → array filled with zeros
arr2 = np.zeros((2, 3))
print("zeros():\n", arr2)

# 3. ones() → array filled with ones
arr3 = np.ones((3, 2))
print("ones():\n", arr3)
```

```python
# 4. full() → array filled with a specific value
arr4 = np.full((2,7),7)
print("full():\n", arr4)

# 5. arange() → evenly spaced values within a range
arr5 = np.arange(0,10,2) #start = 0 , stop = 10, step = 2
print("arange():", arr5)

# 6. linspace() → evenly spaced values between two numbers
arr5 = np.linspace(0,1,5) #5 values between 0 and 1
print("linspace():", arr5)

# 7. diag() → create diagonal matrix
arr7 = np.diag([1,2,3])
print("diag():\n", arr7)

# 8. eye() → create matrix with ones on diagonal
arr8 = np.eye(3)
print("eye():\n", arr8) # eye() being a square

arr = np.eye(3,5)
print("eye():\n", arr) # eye() being a rectangle

# 9 identity() → create identity matrix : diagonal with ones and rest zeros
arr9 = np.identity(3) #identity()can only be square
print("identity():\n", arr9)

### eye() shape can be rectangle or square but identity() shape can only be square

# 10. random.rand() → random values between 0 and 1
arr10 = np.random.rand(2, 3)
print("random.rand():\n", arr10)
```
