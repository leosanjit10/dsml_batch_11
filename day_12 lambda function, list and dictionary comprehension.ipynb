```python
# if there is "e" in a word then print the whole word in capital letter otherwise skip the word
data = ["welcome","to","dsml","course"]
for word in data:
  if "e" in word:
    print(word.upper())
```

```python
#list comprehension
#[expression   for item in iterable    if condition]
 [word.upper() for word in data if "e" in word]
```

```python
# question: input [1,2,3,4,5,6]
#output : ["odd","even","odd","even","odd","even",]
data = [1,2,3,4,5,6]
for number in data:
  if number%2==0:
    print("even")
else:
      print("odd")
      print(data)
```

```python
#upper questions' answer in list comprehension
data = [1,2,3,4,5,6]
result = ["even" if number%2==0 else "odd" for number in data]
print(result)
```

```python
#multiple list comprehension
list1 = [1,2]
list2 = ['A','B']
pairs = [(x,y)for x in list1 for y in list2]
print(pairs)
```

```python
#dictionary comprehension
#syntax -- {key:value  for  vars  in  iterable}
{num:num*num for num in range(1,11)}
```

```python
# dictionary comprehension for getting "odd" "even"
data = [1,2,3,4,5,6]
{num: num**2 for num in data if num %2==0}
```

```python
# conditional dictionary comprehension
# question print even numbers only skip odds
original_dict = {'ram':38,'sita':47,'gita':58,'john':33}
even_dict={(x,y) for x,y in original_dict.items() if y%2==0}
print(even_dict)
```

```python
#multiple conditional dictionary comprehension
original_dict = {'ram':38,'sita':47,'gita':58,'john':33}
new_dict={x:y for x,y in original_dict.items() if y%2 !=0 if y<40}
print(new_dict)
```
