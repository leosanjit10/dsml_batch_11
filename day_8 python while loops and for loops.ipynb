```python
# sum of 10 numbers
sum = 0
i = 1 #(i means to initialize the sum from 1)
while i<=10:
  sum = sum + i
  i = i + 1 #(i += 1 is also same as i = i + 1)
print("sum of 10 numbers",sum)
```

Here’s a clear line-by-line breakdown of your code:

Code Explanation
python
sum = 0
Initializes a variable sum with value 0.

This will hold the total as numbers are added.

python
i = 1  #(i means to initialize the sum from 1)
Sets the variable i to 1.

i acts as a counter, starting from 1.

python
while i <= 10:
Starts a while loop that runs as long as i is less than or equal to 10.

This ensures the loop executes exactly 10 times (for numbers 1 through 10).

python
  sum = sum + i
Adds the current value of i to sum.

For example, first iteration: sum = 0 + 1 = 1, then sum = 1 + 2 = 3, and so on.

python
  i = i + 1  #(i += 1 is also same as i = i + 1)
Increases i by 1 after each loop iteration.

This moves the counter forward so the loop eventually ends.

Shortcut: i += 1 does the same thing.

python
print("sum of 10 numbers", sum)
After the loop finishes, prints the final result.

Output will be:

Code
sum of 10 numbers 55
How It Works Step by Step
Iteration 1: i=1, sum becomes 0+1=1

Iteration 2: i=2, sum becomes 1+2=3

Iteration 3: i=3, sum becomes 3+3=6

… continues until i=10

Final: sum = 55

✅ So the program calculates the sum of the first 10 natural numbers (1 through 10).
