```python
# zero division error
def divide_numbers(num1,num2):
  try:
    result = num1/num2
    print(f"result of {num1}/{num2} is: {result}")
  except ZeroDivisionError:
    print("error: division by zero is not allowed")
try:
  result = 10/0
except ZeroDivisionError:
  print("error: division by zero is not allowed")
```

```python
divide_numbers(10,2)
divide_numbers(10,0)
```

```python
#class task
#try , except, finally , #file upload
def upload_and_read_file(filename):
    try:
        # Try to open the file in read mode
        file = open(filename, "r")
        content = file.read()
        print("File content successfully read:")
        print(content)
    except FileNotFoundError:
        # If the file doesn't exist, handle the error
        print("Error: The file was not found.")
    except PermissionError:
        # If you don’t have permission to open the file
        print("Error: Permission denied to read the file.")
    finally:
        # This block always runs, whether an error occurred or not
        print("File handling attempt finished.")
        try:
            file.close()
            print("File closed successfully.")
        except:
            # If file was never opened, closing will fail
            print("No file to close.")

# Example usage:
upload_and_read_file("example.txt")
```

```python
#raise exception
def check(num):
  if num<0:
    raise ValueError("num cannot be negative")
  print("number",num)
```

```python
try:
  check(-5)
except ValueError as e:
  print(e)
```
