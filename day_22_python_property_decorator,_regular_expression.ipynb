```python
class student:
  def __init__(self,name,age):
    self._name = name # Use internal variable
    self._age = age   # Use internal variable

  @property
  def name(self):
    return self._name # Return internal variable

  @name.setter
  def name(self,value):
    if len(value) > 6:
      raise ValueError("name cannot have more than six characters")
    self._name = value # Set internal variable

  @name.deleter
  def name(self):
    print("deleting name")
    del self._name # Delete internal variable

  @property # Missing @property decorator for the age getter
  def age(self):
    return self._age # Corrected from age.self

  @age.setter
  def age(self,value):
    if value < 0 :
      raise ValueError("age cannot be negative") # Corrected typo: valueError -> ValueError
    self._age = value # Set internal variable

  @age.deleter
  def age(self):
    print("deleting age")
    del self._age # Delete internal variable
```

```python
s = student("Ramesh", 20)

# Getter (access properties)
print("Getter name:", s.name)   # calls getter
print("Getter age:", s.age)

# Setter (modify properties)
s.name = "Sita"                 # valid (<= 6 chars)
s.age = 25
print("Setter name:", s.name)
print("Setter age:", s.age)

# Deleter (delete properties)
del s.name
del s.age
```

```python
# regular expressions (regex)                 # re.match only match the 1st word of the string
import re
text = '''python is awesome.b@gmail.com i love python because python is easy to understand.
you can use python for ai, web development and data science'''
match_result = re.match(r"python", text)
print("match result: ", match_result.group(0) if match_result else "no match")
```

```python
# regular expressions (regex)
import re
text = '''python is awesome.b@gmail.com i love python because python is easy to understand.
you can use python for ai, web development and data science'''
search_result = re.search(r"b@gmail.com", text)
print("search result: ", search_result.group(0) if search_result else "no match")
findall_result = re.findall(r"b@gmail.com", text)
print("findall result: ", findall_result)
```

```python
import re
text = """A proposal to introduce an evening jeep safari in the biologically sensitive Barandabhar forest corridor has drawn strong opposition from conservationists, who warn it could disrupt wildlife movement in one of Nepal’s most critical biological linkages

Jeep safaris are a long-established tourism activity in Chitwan, allowing visitors to observe rare wildlife. At present, safaris operate only in the morning and afternoon, with vehicles required to exit forest areas before sunset.



https://kathmandupost.com/province-no-3/2026/07/03/evening-safari-proposed-to-reap-more-benefits-from-natural-resources-conservationists-warn-against-disturbing-animals-at-rest"""

#re.match() ; match at begining of string
match_result = re.match(r"A proposal", text)
print("re.match():", match_result.group() if match_result else "no match")

#re.search() : find first occurence anywhere
search_result = re.search(r"established", text)
print("re.search():", search_result.group() if search_result else "no match")

#re.findall(): return all matches
findall_result = re.findall(r"safaris", text)
print("re.findall():", findall_result)

#re.finditer : return match objects
print("re.finditer():")
for match in re.finditer(r"a",text):
  print(f"found'{match.group()} at position {match.start()}")

#re.sub() : replace matches
sub_result = re.sub(r"are","were",text)
print("re.sub():",sub_result)

#re.split() : split string
split_result = re.split(r"\.",text)
print("\nre.split():")
for part in split_result:
  print(part.strip())
```
