Create a BankAccount class with:
A public account holder name.
A protected account type.
A private PIN and balance.
Create a SavingAccount class that adds 5% interest to the
balance using methods

```python
class BankAccount:
  def __init__(self,account_holder,account_type,pin,balance):
    self.account_holder = account_holder
    self._account_type = account_type
    self.__pin = pin
    self.__balance = balance
  def get_balance(self,pin):
    if pin == self.__pin:
     return self.__balance
    else:
      return "Invalid Pin"
  def _update_balance(self,amount):
    self.__balance += amount
class SavingAccount(BankAccount):
  def add_interest(self):
    balance = self.get_balance(self.pin)
    if isinstance(balance,(int,float)):
      interest = balance * 0.05
      self._update_balance(interest)
      print(f"interest added: {interest}")

#take input from user
name = input("enter your name: ")
balance = float(input("enter your balance: "))
pin = int(input("set your pin: "))

#create object
account = SavingAccount(name,"Savings",pin,balance)
 #store pin for interest calculation
account.pin = pin

print("--- Account Details ---")
print("Name:" , account.account_holder)
print("Account Type: ", account._account_type)

account.add_interest()
entered_pin = int(input("enter your pin to check balance: "))
print("Balance: ", account.get_balance(entered_pin))
```

Create an abstract BankAccount class with abstract
methods deposit() and withdraw(). Implement them in a
SavingAccount class

```python
from abc import ABC, abstractmethod
class BankAccount(ABC):
  @abstractmethod
  def deposit(self,amount):
    pass
  @abstractmethod
  def withdraw(self,amount):
    pass
class SavingAccount(BankAccount):
  def __init__(self, balance = 0):
    self.balance = balance
  def deposit(self,amount):
    self.balance += amount
    print(f"Deposited: {amount}")
  def withdraw(self, amount):
    if amount > self.balance:
     print("Insufficient Balance")
    else:
      self.balance -= amount
      print(f"Withdrawn : {amount}")
  def check_balance(self):
    print(f"Current Balance: {self.balance}")
account = SavingAccount(10000)
account.deposit(5000)
account.withdraw(2000)
account.check_balance()

```
