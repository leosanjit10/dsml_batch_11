```python
#python recursion
def fact(n):
  if n==0 or n==1:
    return 1
  else:
    return n*fact(n-1)
fact(5)
# the above function is recursion because fact is defined as name for the function and its been called again "fact" in return , i.e within the function
```

```python
#class task
def recur_fib(n):
  if n<=1:
    return n
  else:
      return recur_fib(n-1)+recur_fib(n-2)
      nterms = 10
for i in range(nterms):
  print(recur_fib(i))
```

```python
#nested function : using function inside the function
def outer(x):
  def inner(y):
    return x+y
  return inner
add_five = outer(5)
result = add_five(6)
print(result)
```

```python
# function as a argument
def add(x,y):
  return x+y
def calculate(func,x,y):
  return func(x,y)
result = calculate(add,4,6)
print(result)
```

```python
# return a function as a value
def greeting(name):
  def hello():
    return "hello," + name + "!"
  return hello
greet = greeting("sanjit")
print(greet(()))
```

```python
def mywrapper(func):
  def myfunc():
    print("inside wrapper!")
    func()
  return myfunc
```

```python
# continued from above cell
@mywrapper #(decorator)
def myfunc():
  print("hello, world")
myfunc()
```

```python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"calling {func.__name__}")
        return func(*args, **kwargs)
    return wrapper
```

```python
@logger
def greet(name,age):
  print(f"hello {name},age:{age}")
greet("sanjit",22)
```
