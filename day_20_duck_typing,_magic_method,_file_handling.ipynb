```python
#duck typing
class duck:
  def swim(self):
    return "duck swimming"
  def fly(self):
    return"duck is flying"
class airplane:
  def fly(self):
    return "airplane is flying"
def fly_test(entity):
    print(entity.fly())

duck = duck()
airplane = airplane()
fly_test(duck)
fly_test(airplane)
```

create classes esewa, khalti and banktransfer , each having a pay(amount) method. write a payment processor functiom that works with anu payment method object

```python
class esewa:
  def pay(self, amount):
    print(f"esewa payment of Rs. {amount}")

class khalti:
  def pay(self, amount):
    print(f"khalti payment of Rs. {amount}")

class banktransfer:
  def pay(self, amount):
    print(f"bank transfer of Rs. {amount}")

def payment_processor(entity, amount):
  return entity.pay(amount)

esewa = esewa()
khalti = khalti()
banktransfer = banktransfer()

payment_processor(esewa, 500)
payment_processor(khalti, 700)
payment_processor(banktransfer, 1000)
```

```python
from os import name
#magic method
class bankcustomer:
  def __init__(self,name,age): # here __init__ is magic mrthod
    self.name = name
    self.age = age
  #called by print()
  def __str__(self):
    return f"{self.name} is {self.age} years old"
  #called by repr()
  def __repr__(self):
    return f"bankcustomer('{self.name}','{self.age}')"
  #called by len()
  def __len__(self):
    return self.age
 #makes object callable like a function
  def __call__(self, gender = "male"):
   return f"{self.name} is a {gender}"
#create objects
customer = bankcustomer("sanjit",30)
#__str__
print(customer)
#__repr__
print(repr(customer))
#__len__
print(len(customer))
#__call__
print(customer())
```

Create a BankAccount class and implement __str__()
and __repr__() methods to display account details in
user-friendly and developer-friendly formats.

```python
#class task
class bankaccount:
  def __init__(self,account_id,account_holder):
    self.account_id = account_id
    self.account_holder = account_holder
  def __str__(self):
    return f"{self.account_holder} has a account id { self.account_id}"
  def __repr__(self):
    return f"bank account({self.account_id} , {self.account_holder})"
bankaccount = bankaccount("AMK76145","sanjit")
print(bankaccount)
print(repr(bankaccount))
```

```python
#Changing json to dictionary
import json

# JSON string
json_person = '{"name": "Alice", "age": 25, "city": "Kathmandu"}'

# Convert JSON to dictionary
dict_person = json.loads(json_person)

print(dict_person)
print(type(dict_person))  # <class 'dict'>

```

```python
#changing dictionary to json
import json

# Python dictionary
dict_data = {"name": "Bob", "age": 30, "city": "Pokhara"}

# Convert dictionary to JSON
json_data = json.dumps(dict_data)

print(json_data)
print(type(json_data))  # <class 'str'>
```
