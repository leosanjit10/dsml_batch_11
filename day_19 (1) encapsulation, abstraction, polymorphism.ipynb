Create a BankAccount class with:
A public account holder name.
A protected account type.
A private PIN and balance.
Create a SavingAccount class that adds 5% interest to the
balance using methods

```python
class BankAccount:
  def __init__(self,account_holder,account_type,pin,balance):
    self.account_holder = account_holder
    self._account_type = account_type
    self.__pin = pin
    self.__balance = balance
  def get_balance(self,pin):
    if pin == self.__pin:
     return self.__balance
    else:
      return "Invalid Pin"
  def _update_balance(self,amount):
    self.__balance += amount
class SavingAccount(BankAccount):
  def add_interest(self):
    balance = self.get_balance(self.pin)
    if isinstance(balance,(int,float)):
      interest = balance * 0.05
      self._update_balance(interest)
      print(f"interest added: {interest}")

#take input from user
name = input("enter your name: ")
balance = float(input("enter your balance: "))
pin = int(input("set your pin: "))

#create object
account = SavingAccount(name,"Savings",pin,balance)
 #store pin for interest calculation
account.pin = pin

print("--- Account Details ---")
print("Name:" , account.account_holder)
print("Account Type: ", account._account_type)

account.add_interest()
entered_pin = int(input("enter your pin to check balance: "))
print("Balance: ", account.get_balance(entered_pin))
```

Create an abstract BankAccount class with abstract
methods deposit() and withdraw(). Implement them in a
SavingAccount class

```python
from abc import ABC, abstractmethod
class BankAccount(ABC):
  @abstractmethod
  def deposit(self,amount):
    pass
  @abstractmethod
  def withdraw(self,amount):
    pass
class SavingAccount(BankAccount):
  def __init__(self, balance = 0):
    self.balance = balance
  def deposit(self,amount):
    self.balance += amount
    print(f"Deposited: {amount}")
  def withdraw(self, amount):
    if amount > self.balance:
     print("Insufficient Balance")
    else:
      self.balance -= amount
      print(f"Withdrawn : {amount}")
  def check_balance(self):
    print(f"Current Balance: {self.balance}")
account = SavingAccount(10000)
account.deposit(5000)
account.withdraw(2000)
account.check_balance()

```

Runtime polymorphism occurs
when a subclass provides its own
specific implementation of a
method that is already defined in
its parent class. Python resolves
these method calls dynamically at
runtime based on the object
calling it

```python
# polymorphism in python
  # : runtime polymorphism (method overriding)  # here process_payment method is overrriden in subclasses, thats why execute_payment func is called. this is runtime polymorphim
class paymentgateway:
  def process_payment(self,amount):
    pass
class creditcard(paymentgateway):
  def process_payment(self,amount):
    return f"processing ${amount} via credit card authentication"
class paypal(paymentgateway):
  def process_payment(self,amount):
    return f"processing ${amount} via paypal secure token"
#united interface function
def execute_checkout(payment_method,amount):
 print(payment_method.process_payment(amount))
#usage
card = creditcard()
wallet = paypal()
execute_checkout(card,250)
execute_checkout(wallet,250)
```

Duck typing is a concept related to dynamic
typing where the type or class of an object is
less important than the methods it defines.
The name comes from the saying: “If it walks
like a duck and quacks like a duck, then it
probably is a duck.

```python
#polymorphism:
   #Duck typing
class duck:
  def swim(self):
    return "duck swimming"
  def fly(self):
    return "duck flying"
class airplane:
  def fly(self):
    return "aeroplane flying"
def fly_test(entity):
  print(entity.fly())
#create objects
duck = duck()
airplane = airplane()
#same functions works with different objects
fly_test(duck)
fly_test(airplane)
```
