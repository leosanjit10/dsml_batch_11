```python
def calculator(a,b, operator):
  if operator == "+":
      return a + b
  elif operator == "-":
       return a - b
  elif operator == "*":
        return a * b
  elif operator == "/":
        return a / b
a = float(input("enter a number"))
b = float(input('enter second number'))
operator = input('enter an operator(+,-,*,/)')
print(calculator(a,b, operator))
```

```python
def list_operation():
  numbers = []
  for i in range(5):
    value = input(f"enter value {i+1}:")
    if "." in value:
      if len(numbers)>0:
        removed = numbers.pop()
        print(f"float detected. Removed: {removed}")
      else:
        print("float detected, but the list is empty")
    else:
        numbers.append(int(value))
        print(f"integer{value}added to the list")
    print("updated list:", numbers)
list_operation()


```
