```python
# create a polygon class and use it to represent a triangle, square, pentagon,hexagon
class polynomial:
  def triangle(self):
    print("this is a triangle")
  def square(self):
    print("this is a square")
  def pentagon(self):
    print("this is a pentagon")
  def hexagon(self):
    print("this is a hexagon")
```

```python
obj = polynomial()
obj.triangle()
```

```python
# method overloading --  multiple methods in the same class share the same name but differ in their parameter lists (number, type, or order of parameters
class calculator:
  def add(self,a,b):
    return(a+b)
  def add(self,a,b,c):
     return(a+b+c)
```

```python
obj = calculator()
obj.add(2,3)
```

```python
obj = calculator()
obj.add(2,3,5)
```

```python
class calculator:  #here we did method overloading by passing value of c as 0
  def add(self,a,b):
   return a+b
  def add(self,a,b,c = 0):
   return a+b+c
```

```python
obj2 = calculator()
obj2.add(2,3)
```

```python
# another way of method overloading is using*args to pass multiple arguments
class calculator:
  def add(self,*args):
    return sum(args)
```

```python
obj4 = calculator()
obj4.add(2,3,4,5,6)
```

```python
obj5 = calculator()
obj5.add(2,3,4,5,6,7,8,9)
```

```python
# inheritance
   #types--single inheritance
class Animal:
  def sound(self):
    print("Animals makes sound")
class Dog(Animal):
  def bark(self):
    print("Dog barks")


```

```python
d = Dog()
d.sound()
d.bark()
```

```python
#multiple inheritance -- two parent class, 1 child class
class father:
  def skills(self):
    print("gardening")
class mother:
  def talents(self):
    print("cooking")

class child(father,mother):
  pass
```

```python
c = child()
c.skills()
c.talents()
```

```python
# multilevel inheritance
class grandparent:
  def skills(self):
    print("lawn mowing")
class father(grandparent):
  def talents(self):
    print("gardening")
class child(father):
  def character(self):
    print("happy")

```

```python
obj = child()
obj.skills()
obj.talents()
obj.character()
```

```python
#hierarchical inheritance -- 1 parent class, multiple child class
class father:
  def skills(self):
    print("gardening")
class child1(father):
  def talents(self):
    print("cooking")
class child2(father):
  def character(self):
    print("happy")
```

```python
obj1 = child1()
obj2 = child2()
obj1.skills()
obj1.talents()
obj2.skills()
obj2.character()
```

```python
class ottpsubscription: #parent class
  def __init__(self,subscription_id,plan,total_payment):
    self.id = subscription_id
    self.plan = plan
    self.total_payment = total_payment
  def subscription(self):
    print(f"subscriber with{self.id} id subscribed to the {self.plan}")
  def unsubscribe(self):
    print(f"subscriber with {self.id} unsubscribed to the {self.plan}")


```

```python
class premiumsubscription(ottpsubscription):
  def __init__(self,subscription_id,plan,total_payment,screens):
    super().__init__(subscription_id,plan,total_payment)
    self.max_screens = screens
  def set_max_screens(self, screens):
    self.max_screens = screens
    print(f"maximum screen set to {self.max_screens} in the premium plan")
```

```python
obj = premiumsubscription(986754, "pro",8000,4)
```

```python
obj.set_max_screens(9)
```

```python
obj.subscription()
```

```python
obj.unsubscribe()
```

```python
class person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def show_details(self):
    print(f"name:{self.name}, age: {self.age}")
class student(person):
  def __init__(self,name,age,student_id):
    person.__init__(self,name,age) # Corrected: Directly call Person's __init__
    self.student_id = student_id
  def study(self):
    print(f"student {self.name} with ID {self.student_id} is studying")
class teacher(person):
 def __init__(self,name,age,subject):
  super().__init__(name,age)
  self.subject = subject
 def teach(self):
   print(f"teacher {self.name} is teaching {self.subject}")
class TeachingAssistant(student,teacher):
   def __init__(self,name,age,student_id,subject,department,salary):
    student.__init__(self,name,age,student_id)
    teacher.__init__(self,name,age,subject)
    self.department = department
    self.salary = salary
   def assist(self):
    print(f"Teaching Assistant {self.name} assists in {self.subject}",
          f"belongs to {self.department} department",
          f"and earns a salary of {self.salary}.")

obj = TeachingAssistant("sanjit", 30, 123456, "dsml", "CS", "100k")
```

```python
obj.show_details()
obj.study()
obj.teach()
obj.assist()
```
